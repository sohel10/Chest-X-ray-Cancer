```python
from dataclasses import dataclass
from pathlib import Path
```

```python

@dataclass(frozen=True)
class DataIngestionConfig:
    root_dir: Path
    source_URL: str
    local_data_file: Path
    unzip_dir: Path
```

```python
from cnnClassifier.constants import *
from cnnClassifier.utils.common import read_yaml, create_directories
```

```python
class ConfigurationManager:
    def __init__(
        self,
        config_filepath = CONFIG_FILE_PATH,
        params_filepath = PARAMS_FILE_PATH):

        self.config = read_yaml(config_filepath)
        self.params = read_yaml(params_filepath)

        create_directories([self.config.artifacts_root])


    
    def get_data_ingestion_config(self) -> DataIngestionConfig:
        config = self.config.data_ingestion

        create_directories([config.root_dir])

        data_ingestion_config = DataIngestionConfig(
            root_dir=config.root_dir,
            source_URL=config.source_URL,
            local_data_file=config.local_data_file,
            unzip_dir=config.unzip_dir 
        )

        return data_ingestion_config
```

```python
import os
import zipfile
from cnnClassifier import logger

class DataIngestion:
    def __init__(self, config):
        self.config = config

    def extract_zip_file(self):
        """
        Extracts dataset ZIP file safely.
        """

        zip_path = self.config.local_data_file         # e.g. artifacts/data_ingestion/data_ct.zip
        extract_path = self.config.unzip_dir           # e.g. artifacts/data_ingestion/

        # 1️⃣ Check if the ZIP exists
        if not os.path.exists(zip_path):
            raise FileNotFoundError(f"ZIP file missing: {zip_path}")

        logger.info(f"Extracting ZIP file: {zip_path}")

        # 2️⃣ Extract safely
        try:
            with zipfile.ZipFile(zip_path, "r") as zip_ref:
                zip_ref.extractall(extract_path)

            logger.info(f"Extraction completed → {extract_path}")

        except zipfile.BadZipFile:
            raise ValueError("ZIP file is corrupted or invalid.")

        except Exception as e:
            raise e
```

```python
from cnnClassifier.config.configuration import ConfigurationManager
from cnnClassifier.components.data_ingestion import DataIngestion

config = ConfigurationManager()
cfg = config.get_data_ingestion_config()

data = DataIngestion(cfg)
data.download_file()
data.extract_zip_file()
```
